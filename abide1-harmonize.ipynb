```python
import pandas as pd
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
```

```python
df = pd.read_csv('abide-structural.csv')
```

```python
df.head()
```

```python
def search_keys(filter):
    found = []
    for key in df.keys():
        result = []
        for val in filter:
            if val[0] is 'or':
                result.append(any([item in key.lower() for item in val[1]]))
            if val[0] is 'and':
                result.append(all([item in key.lower() for item in val[1]]))
        if all(result):
            found.append(key)
    return found
```

```python
vol_keys = search_keys([('and', ['vol'])])
vol_keys
```

```python
df_ilx['ID_Unnamed:_5_level_1'] = df_ilx['ID_Unnamed:_5_level_1'].apply(lambda x: int(x) if not pd.np.isnan(x) else -1)
```

```python
names1 = list(df_ilx['Atlas_Segmentation_Label_Unnamed:_6_level_1'])
names1 = [str(name).replace('-', ' ').replace('_', ' ') for name in names1]
names2 = list(df_ilx.Federated_DE_Name)
```

```python
from collections import namedtuple
Info = namedtuple('Info', "iri, label, source")
```

```python
def find_ilx(label, source, debug=False):
    if debug: 
        print(label)
    if isinstance(label, str):
        result1 = process.extractOne(label, names1,
                                scorer=fuzz.token_sort_ratio)
        if debug: 
            print(result1)
        result2 = process.extractOne(label, names2,
                                scorer=fuzz.token_sort_ratio)
        if debug: 
            print(result2)
        result, names = (result1, names1) if result1[1] > result2[1] else (result2, names2)
        label = df_ilx.Federated_DE_Name[names.index(result[0])]
        if label:
            return Info(iri=df_ilx.index[names.index(result[0])], 
                        label=label,
                        source=source)
    else:
        matches = df_ilx.Federated_DE_Name[df_ilx['ID_Unnamed:_5_level_1'].values == label]
        matches = matches.drop_duplicates().dropna()
        if debug: 
            print(matches)
        if matches.shape[0] > 1:
            raise ValueError('Multiple matches: {}'.format(matches))
        if debug: 
            print(matches.shape)
        if matches.shape[0] > 0:
            return Info(iri=matches.index.values[0], label=matches.values[0],
                    source=source)
    return None
```

```python
known_failures = ['swf_Background_vol', ' BrainSegVol', ' BrainSegVolNotVent',
                  ' BrainSegVolNotVentSurf', ' VentricleChoroidVol',
                  ' SupraTentorialVolNotVent', ' SupraTentorialVolNotVentVox',
                  ' MaskVol']
```

```python
def map_swf(key, debug=False):
    label = ' '.join(key.split('_')[1:]).replace('-', ' ')
    match = find_ilx(label, 'FSL', debug=debug)
    return match

def map_fsglobal(key, debug=False):
    match = find_ilx(key, 'FreeSurfer', debug=debug)
    return match

def map_label(key, debug=False):
    id = int(key.split('_')[2])
    source=None
    if key.startswith('fs'):
        source = 'FreeSurfer'
    if key.startswith('ants'):
        source = 'ANTS'
    match = find_ilx(id, source, debug=debug)
    return match
```

```python
keymap = {}
for key in vol_keys:
    if key in known_failures:
        continue
    if key.startswith('swf'):
        keymap[key] = map_swf(key)
    if key.startswith(' '):
        keymap[key] = map_fsglobal(key)
    if 'label' in key:
        keymap[key] = map_label(key)
    if keymap[key] is None:
        del keymap[key]
```
