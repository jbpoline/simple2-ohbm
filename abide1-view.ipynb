```python
df_pheno = pd.read_csv('Phenotypic_V1_0b.csv', na_values=[-9999])
df_pheno.SUB_ID = df_pheno.SUB_ID.apply(lambda x: 'sub-{:07d}'.format(x))
df_pheno.index = df_pheno.SUB_ID
```

```python
df2_pheno = df_pheno[['SITE_ID', 'DX_GROUP', 'DSM_IV_TR', 'AGE_AT_SCAN', 'SEX',
                      'HANDEDNESS_CATEGORY', 'FIQ', 'VIQ', 'PIQ']]
df2_pheno.head()
```

```python
labels = ['anterior corpus callosum volume', 
          'central corpus callosum volume', 
          'middle anterior corpus callosum volume',
          'middle posterior corpus callosum volume', 
          'posterior corpus callosum volume', 
          'cerebrum tissue volume',
          'left caudate nucleus volume', 
          'left putamen volume',
         ]
data = df2.stack(level=1)[labels].reset_index().dropna() #.stack(level=0).reset_index()
CC_vol = data[['anterior corpus callosum volume', 
          'central corpus callosum volume', 
          'middle anterior corpus callosum volume',
          'middle posterior corpus callosum volume', 
          'posterior corpus callosum volume']].sum(axis=1)
stria_vol = data[['left caudate nucleus volume', 
          'left putamen volume']].sum(axis=1)
stat_df = pd.concat((data[['participant_id', 'cerebrum tissue volume']], 
                     pd.DataFrame(CC_vol, columns=['corpus callosum volume']),
                     pd.DataFrame(CC_vol/data['cerebrum tissue volume'], columns=['scaled corpus callosum volume']),
                     pd.DataFrame(stria_vol, columns=['striatal volume'])), axis=1)
stat_df.index = stat_df.participant_id
stat_df.head()
```

Hypotheses

PIET-1: Total Brain Volume will positively correlate with IQ (in both sexes across the complete age range).

MAC-1: Left striatum volume (caudate + putamen) will positively correlate with IQ in the total (male + female) child (age &lt; 20) group.

MAC-2: Left striatum volume (caudate + putamen) will positively correlate with IQ in the male children group.

MAC-3: Left striatum volume (caudate + putamen) will not correlate with IQ in the female children group.

MAC-3: Left striatum volume (caudate + putamen) will not correlate with IQ in the female children group.

GANJ-1: Total Corpus Callosum midsagittal area, after correcting for total brain volume, will negatively correlate with IQ.

GANJ-2: Total Corpus Callosum midsagittal area, after correcting for total brain volume, will negatively correlate with IQ in the young (age &lt; 12) group.

GANJ-3: Total Corpus Callosum midsagittal area, after correcting for total brain volume, will not significantly correlate with IQ in the adolescent (age &gt; 12) group.

GANJ-4:. Total Corpus Callosum midsagittal area, after correcting for total brain volume, will negatively correlate with IQ in the male (age &lt; 12) group.

GANJ-5: Total Corpus Callosum midsagittal area, after correcting for total brain volume, will not significantly correlate with IQ in the female (age &lt; 12) group.

```python
combo_df = pd.concat((df2_pheno[df2_pheno.DX_GROUP == 2], stat_df), axis=1, sort=False)
combo_df.head()
```

```python
import statsmodels as sm
```

```python
sns.lmplot(x='cerebrum tissue volume', y='FIQ', data=combo_df[['cerebrum tissue volume', 'FIQ']].dropna(),
          fit_reg=True, robust=True);
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats.stats import pearsonr
```

```python
data2 = combo_df[['cerebrum tissue volume', 'FIQ']].dropna()
md = smf.ols("FIQ ~ Q('cerebrum tissue volume')", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['cerebrum tissue volume']))
```

```python
data2 = combo_df[combo_df.AGE_AT_SCAN < 20][['striatal volume', 'FIQ']].dropna()
md = smf.ols("FIQ ~ Q('striatal volume')", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['striatal volume']))
```

```python
data2 = combo_df[combo_df.AGE_AT_SCAN < 20][combo_df.SEX == 1.0][['striatal volume', 'FIQ']].dropna()
md = smf.ols("FIQ ~ Q('striatal volume')", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['striatal volume']))
```

```python
data2 = combo_df[['FIQ', 'scaled corpus callosum volume']].dropna()
md = smf.ols("FIQ ~ Q('scaled corpus callosum volume')", 
             data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['scaled corpus callosum volume']))
```

```python
data2 = combo_df[combo_df.AGE_AT_SCAN < 12][['scaled corpus callosum volume', 'FIQ']].dropna()
md = smf.ols("FIQ ~ Q('scaled corpus callosum volume')", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['scaled corpus callosum volume']))
```

```python
data2 = combo_df[combo_df.AGE_AT_SCAN < 12][combo_df.SEX == 1.0][['scaled corpus callosum volume', 'FIQ']].dropna()
md = smf.ols("FIQ ~ Q('scaled corpus callosum volume')", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['scaled corpus callosum volume']))
```
