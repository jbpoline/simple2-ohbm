```python
labels = ['anterior corpus callosum volume', 
          'central corpus callosum volume', 
          'middle anterior corpus callosum volume',
          'middle posterior corpus callosum volume', 
          'posterior corpus callosum volume', 
          'cerebrum tissue volume',
          'left caudate nucleus volume', 
          'left putamen volume',
         ]
data = df2.stack(level=1)[labels].reset_index().dropna() #.stack(level=0).reset_index()
CC_vol = data[['anterior corpus callosum volume', 
          'central corpus callosum volume', 
          'middle anterior corpus callosum volume',
          'middle posterior corpus callosum volume', 
          'posterior corpus callosum volume']].sum(axis=1)
stria_vol = data[['left caudate nucleus volume', 
          'left putamen volume']].sum(axis=1)
stat_df = pd.concat((data[['participant_id', 'cerebrum tissue volume']], 
                     pd.DataFrame(CC_vol, columns=['corpus callosum volume']),
                     pd.DataFrame(CC_vol/data['cerebrum tissue volume'], columns=['scaled corpus callosum volume']),
                     pd.DataFrame(stria_vol, columns=['striatal volume'])), axis=1)
stat_df.index = stat_df.participant_id
stat_df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats.stats import pearsonr
```

#### both sexes
    - regressed out: site and age

```python
data2 = combo_df[['cerebrum tissue volume', 'FIQ', 'SITE_ID', 'AGE_AT_SCAN', 'SEX']].dropna()
#data2 = data2.loc[data2['SEX']==1.0]

md = smf.ols("FIQ ~ Q('cerebrum tissue volume') + SITE_ID + AGE_AT_SCAN", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['cerebrum tissue volume']))
```

#### hyp.1 females
    - regressed out site and age

```python
data2 = combo_df[['cerebrum tissue volume', 'FIQ', 'SITE_ID', 'AGE_AT_SCAN', 'SEX']].dropna()
data2 = data2.loc[data2['SEX']==1.0]

md = smf.ols("Q('cerebrum tissue volume')  ~ FIQ + SITE_ID + AGE_AT_SCAN", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['cerebrum tissue volume']))
```

```python
data2 = combo_df[['cerebrum tissue volume', 'FIQ', 'SITE_ID', 'AGE_AT_SCAN', 'SEX']].dropna()
data2 = data2.loc[data2['SEX']==2.0]

md = smf.ols("Q('cerebrum tissue volume') ~ FIQ + SITE_ID + AGE_AT_SCAN", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['cerebrum tissue volume']))
```

```python
data2 = combo_df[combo_df.AGE_AT_SCAN >= 18][['cerebrum tissue volume','FIQ','SITE_ID','AGE_AT_SCAN','SEX']].dropna()
md = smf.ols("FIQ ~ Q('cerebrum tissue volume') + SITE_ID + SEX", data=data2)
mdf = md.fit()
print(mdf.summary())
print(pearsonr(data2['FIQ'], data2['cerebrum tissue volume']))
```
